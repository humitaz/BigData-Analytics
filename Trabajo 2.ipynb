En la actualidad contamos con diferentes fuentes de información, que nos permiten realizar investigaciones de todo tipo y a todo nivel; sin embargo, esta creciente variedad es la que podría condicionar la permanencia de ciertas herramientas de recopilación de información. El auge del Big Data, como herramienta que agiliza el procesamiento de grandes bases de datos, es innegable, y ha generado ciertos cuestionamientos en relación a la eficiencia del desarrollo de encuestas. 

En este sentido, tenemos dos situaciones, no necesariamente excluyentes: good data y/o big data. Por su parte, las encuestas son herramientas más cercanas a los “buenos datos” pues son amplias en variables, mientras que el big data, es amplia en cantidad de observaciones. Es por ello, que optar únicamente por encuestas como las conocemos en pleno auge del Big Data, es quedarnos atorados en el tiempo. Mientras que, por otro lado, apostar por la caducidad de las encuestas es minimizar la potencia de estas en cuanto a buenos datos confiables, especialmente cuando se trata de investigación en temas sociales.

En cuanto a las encuestas, estas brindan información relevante sobre variables como: género, edad, ubicación geográfica, etc., las cuales serán representativas a nivel poblacional, por lo cual aspectos como el nivel socioeconómico no condicionarán su inclusión en el análisis. Mientras que, el big data que puede utilizar plataformas como Facebook, Twitter, Mechanical Turk, etc., tienen mayor tendencia a generar sesgos de selección y de medición, principalmente porque no toda la población utiliza estas plataformas. Además, de no ser constantes en el tiempo pues podrían desaparecer. También, podrían modificar sus términos de privacidad, afectando la cantidad de usuarios; e incluso precisamente por temas de confidencialidad, el acceso a toda la información puede llegar a ser limitado. Asimismo, la cantidad de cuentas falsas o duplicadas, generan desconfianza en relación a la información recopilada. 

Por su parte, el big data, a diferencia de las encuestas, es una herramienta que permite facilidad en la toma de decisiones, pues permite el procesamiento rápido de datos. Además, facilita el seguimiento de los sujetos de estudio, identificando tendencias y necesidades, pues se cuenta con información en tiempo real. Asimismo, permite la automatización de procesos, redistribuyendo eficientemente recursos escasos como el tiempo.

Un ejemplo, son las elecciones generales de primera vuelta del 2021 en Perú. Por un lado, las redes sociales como Twitter, evidenciaban cierta preferencia por determinados candidatos, mientras que las encuestas, a pocos días de las elecciones mostraban un escenario diferente y un poco más cercano al real. Este proceso evidenció un escenario bastante obvio para algunos: la mayoría de la población peruana no usa Twitter y/o Facebook. Asimismo, muchas encuestadoras no pueden llegar a espacios de difícil acceso, lo cual resta participación a ciertos grupos determinantes. 

Entonces, las grandes bases de datos están sujetas a la calidad de la información que puedan brindar, siendo el principal reto obtener datos necesarios de manera eficiente. Por ello, ambas herramientas son importantes dependiendo del escenario e incluso podrían llegar a ser complementarias, pero difícilmente una reemplazará a la otra.
